## ¿Qué es el cancer de pulmón?
El cáncer de pulmón es un tipo de tumor maligno que se origina en los pulmones, generalmente asociado al tabaquismo, pero también puede ser causado por factores genéticos o ambientales. Existen dos tipos principales: el cáncer de pulmón de células no pequeñas y el de células pequeñas.

# Síntomas del cáncer de pulmón:
1. Tos persistente o que empeora con el tiempo.
2. Dolor en el pecho que se intensifica al respirar profundamente, toser o reír.
3. Dificultad para respirar (disnea).
4. Sibilancias o voz ronca.
5. Pérdida de peso sin causa aparente.
6. Fatiga constante o debilidad.
7. Infecciones pulmonares recurrentes (bronquitis o neumonía).
8. Tos con sangre (hemoptisis)

# Indicadores del cáncer de pulmón para personas no médicas
1. Tos persistente.
2. Dificultad para respirar o sensación de falta de aire.
3. Dolor en el pecho al respirar, toser o reír.
4. Tos con sangre o esputo con color anormal.
5. Ronquera o cambios en la voz sin causa aparente.
6. Pérdida de peso inexplicada.
7. Cansancio extremo o debilidad constante.
8. Infecciones pulmonares frecuentes (bronquitis o neumonía repetidas).
9. Hinchazón en la cara o el cuello.
10. Dolor en los huesos o en otras partes del cuerpo sin explicación.

# Preprocesamiento para la clasificación:

```python
x = data.drop("LUNG CANCER", axis = 1)
y = data["LUNG CANCER"]
column_names = data.drop("LUNG CANCER", axis=1).columns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
x = scaler.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

print(f"Shape of training data : {x_train.shape}, {y_train.shape}")
print(f"Shape of testing data : {x_test.shape}, {y_test.shape}")
```

# Modelo de Regresión Logística:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
lr_conf = confusion_matrix(y_test, lr_pred)
lr_report = classification_report(y_test, lr_pred)
lr_acc = round(accuracy_score(y_test, lr_pred)*100, ndigits = 2)
print(f"Confusion Matrix : \n\n{lr_conf}")
print(f"\nClassification Report : \n\n{lr_report}")
print(f"\nThe Accuracy of Logistic Regression is {lr_acc} %")


```

```python
# Obtener los coeficientes
feature_importance = pd.Series(lr.coef_[0], index= column_names).sort_values(ascending=False)

# Mostrar los coeficientes
print(feature_importance)

# Graficar la importancia de cada variable
plt.figure(figsize=(10, 5))
feature_importance.plot(kind="bar", color=palette[11])
plt.title("Importancia de las Variables en la Regresión Logística")
plt.xlabel("Síntomas")
plt.ylabel("Coeficiente")
plt.show()

```

## Supuestos:
- Relación Lineal entre las Variables Independientes y el Logit
- No Colinealidad entre Variables Predictoras
- Independencia de Errores
- Ausencia de Valores Extremos o Outliers Influyentes

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.stattools import durbin_watson

# Cargar el dataset
df = pd.read_csv("survey lung cancer.csv")

# Convertir variables categóricas a numéricas si no lo están
df["LUNG_CANCER"] = df["LUNG_CANCER"].map({'YES': 1, 'NO': 0})
df["SMOKING"] = df["SMOKING"].astype(int)
df["ALCOHOL CONSUMING"] = df["ALCOHOL CONSUMING"].astype(int)
df["COUGHING"] = df["COUGHING"].astype(int)
df["SHORTNESS OF BREATH"] = df["SHORTNESS OF BREATH"].astype(int)
df["FATIGUE "] = df["FATIGUE "].astype(int)

# Definir variables independientes y dependientes
X = df[['AGE', 'SMOKING', 'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'FATIGUE ']]
X = sm.add_constant(X)  # Agregar intercepto
y = df['LUNG_CANCER']

# Ajustar modelo de regresión logística con statsmodels
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# Obtener los residuos del modelo ajustado
residuals = result.resid_response  

# Histograma de los residuos
plt.figure(figsize=(8,5))
sns.histplot(residuals, kde=True, bins=30)
plt.title("Distribución de los residuos")
plt.xlabel("Residuos")
plt.ylabel("Frecuencia")
plt.show()

# Gráfico Q-Q de los residuos
plt.figure(figsize=(8,5))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Gráfico Q-Q de los residuos")
plt.show()

# Prueba de normalidad de Shapiro-Wilk
shapiro_test = stats.shapiro(residuals)
print(f"Prueba de Shapiro-Wilk: estadístico={shapiro_test.statistic:50f}, p-valor={shapiro_test.pvalue:.50f}")

# Prueba de Kolmogorov-Smirnov
ks_test = stats.kstest(residuals, 'norm', args=(residuals.mean(), residuals.std()))
print(f"Prueba de Kolmogorov-Smirnov: estadístico={ks_test.statistic:.50f}, p-valor={ks_test.pvalue:.50f}")

# Estadístico de Durbin-Watson para autocorrelación de residuos
dw_stat = durbin_watson(residuals)
print(f"Estadístico de Durbin-Watson: {dw_stat:.2f}")

# Interpretación de normalidad
if shapiro_test.pvalue > 0.05 and ks_test.pvalue > 0.05:
    print("Los residuos siguen una distribución normal (No se rechaza la hipótesis nula).")
else:
    print("Los residuos NO siguen una distribución normal (Se rechaza la hipótesis nula).")
```

```python
# SUPUESTOS DE REGRESIÓN LOGÍSTICA
independent_vars = ['AGE', 'SMOKING', 'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'FATIGUE ']
X = df[independent_vars]
X = sm.add_constant(X)
Y = df['LUNG_CANCER']

# 1. Verificación de la relación lineal con el logit mediante Box-Tidwell
X['log_AGE'] = np.log(X['AGE'] + 1) * X['AGE']
logit_model = sm.Logit(Y, X)
result = logit_model.fit(disp=0)

if result.pvalues['log_AGE'] < 0.05:
    print("No se cumple el supuesto de linealidad en logit (Box-Tidwell).")
else:
    print("Se cumple el supuesto de linealidad en logit.")

# 2. Verificación de multicolinealidad con VIF
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("\nFactor de Inflación de Varianza (VIF):")
print(vif_data)
if vif_data['VIF'].max() > 10:
    print("Existe multicolinealidad en las variables independientes.")
else:
    print("No hay problemas graves de multicolinealidad.")

# 3. Identificación de valores atípicos mediante distancia de Cook
influence = result.get_influence()
cooks_d = influence.cooks_distance[0]

plt.figure(figsize=(8,4))
plt.stem(cooks_d, markerfmt=",")  
plt.axhline(4/X.shape[0], color='r', linestyle='dashed', label="Umbral (4/n)")
plt.title("Distancia de Cook")
plt.xlabel("Observaciones")
plt.ylabel("Distancia de Cook")
plt.legend()
plt.show()

if any(cooks_d > (4/X.shape[0])):
    print("Existen valores atípicos que pueden influir en el modelo.")
else:
    print("No hay valores atípicos con alta influencia en el modelo.")
```

### Conclusión de los supuestos:
- Linealidad en el logit: Se cumple, ya que la prueba de Box-Tidwell no indicó problemas de linealidad.  
- Multicolinealidad: No se cumple, ya que hay variables con un VIF extremadamente alto (especialmente AGE y log_AGE), lo que sugiere una fuerte correlación entre ellas.  
- Valores atípicos: No se cumple, pues hay observaciones con una distancia de Cook superior al umbral, lo que indica que ciertos puntos tienen una alta influencia en el modelo.

## Comparación entre la edad de los fumadores y los no fumadores.

```python
# Comparar la edad entre fumadores y no fumadores
group_non_smokers = df[df["SMOKING"] == 1]["AGE"]
group_smokers = df[df["SMOKING"] == 2]["AGE"]

# Prueba ANOVA
f_stat, p_value = stats.f_oneway(group_non_smokers, group_smokers)

# Resultados
print(f"Estadístico F: {f_stat:.2f}")
print(f"Valor p: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusión: Existen diferencias significativas en la edad entre fumadores y no fumadores.")
else:
    print("Conclusión: No hay diferencias significativas en la edad entre fumadores y no fumadores.")
```

## ¿Qué es el cancer de pulmón?
El cáncer de pulmón es un tipo de tumor maligno que se origina en los pulmones, generalmente asociado al tabaquismo, pero también puede ser causado por factores genéticos o ambientales. Existen dos tipos principales: el cáncer de pulmón de células no pequeñas y el de células pequeñas.

# Síntomas del cáncer de pulmón:
1. Tos persistente o que empeora con el tiempo.
2. Dolor en el pecho que se intensifica al respirar profundamente, toser o reír.
3. Dificultad para respirar (disnea).
4. Sibilancias o voz ronca.
5. Pérdida de peso sin causa aparente.
6. Fatiga constante o debilidad.
7. Infecciones pulmonares recurrentes (bronquitis o neumonía).
8. Tos con sangre (hemoptisis)

# Indicadores del cáncer de pulmón para personas no médicas
1. Tos persistente.
2. Dificultad para respirar o sensación de falta de aire.
3. Dolor en el pecho al respirar, toser o reír.
4. Tos con sangre o esputo con color anormal.
5. Ronquera o cambios en la voz sin causa aparente.
6. Pérdida de peso inexplicada.
7. Cansancio extremo o debilidad constante.
8. Infecciones pulmonares frecuentes (bronquitis o neumonía repetidas).
9. Hinchazón en la cara o el cuello.
10. Dolor en los huesos o en otras partes del cuerpo sin explicación.

# Preprocesamiento para la clasificación:

```python
x = data.drop("LUNG CANCER", axis = 1)
y = data["LUNG CANCER"]
column_names = data.drop("LUNG CANCER", axis=1).columns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
x = scaler.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

print(f"Shape of training data : {x_train.shape}, {y_train.shape}")
print(f"Shape of testing data : {x_test.shape}, {y_test.shape}")
```

# Modelo de Regresión Logística:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
lr_conf = confusion_matrix(y_test, lr_pred)
lr_report = classification_report(y_test, lr_pred)
lr_acc = round(accuracy_score(y_test, lr_pred)*100, ndigits = 2)
print(f"Confusion Matrix : \n\n{lr_conf}")
print(f"\nClassification Report : \n\n{lr_report}")
print(f"\nThe Accuracy of Logistic Regression is {lr_acc} %")


```

```python
# Obtener los coeficientes
feature_importance = pd.Series(lr.coef_[0], index= column_names).sort_values(ascending=False)

# Mostrar los coeficientes
print(feature_importance)

# Graficar la importancia de cada variable
plt.figure(figsize=(10, 5))
feature_importance.plot(kind="bar", color=palette[11])
plt.title("Importancia de las Variables en la Regresión Logística")
plt.xlabel("Síntomas")
plt.ylabel("Coeficiente")
plt.show()

```

# Análisis de los supuestos:

## Supuestos:
- Relación Lineal entre las Variables Independientes y el Logit
- No Colinealidad entre Variables Predictoras
- Independencia de Errores
- Ausencia de Valores Extremos o Outliers Influyentes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
from scipy.stats import chi2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular probabilidades predichas
y_pred_prob = lr.predict_proba(x_train)[:, 1]

# Transformación logit
logit = np.log(y_pred_prob / (1 - y_pred_prob))

# Crear DataFrame con los valores
df_logit = pd.DataFrame({'Logit': logit})

# Agregar las variables predictoras al DataFrame
for i, col in enumerate(column_names):
    df_logit[col] = x_train[:, i]

# Ajustar el número de columnas para evitar el error
n_cols = np.ceil(len(column_names) / 2).astype(int)
fig, axes = plt.subplots(nrows=2, ncols=n_cols, figsize=(15, 10))
axes = axes.flatten()

# Graficar la relación Logit vs cada variable
for i, col in enumerate(column_names):
    sns.scatterplot(x=df_logit[col], y=df_logit['Logit'], ax=axes[i], color='yellow', edgecolor='black')
    axes[i].set_title(f'Logit vs {col}', color="black")
    axes[i].set_xlabel(col, color="black")
    axes[i].set_ylabel("Logit", color="black")
    axes[i].tick_params(colors="black")

# Ocultar los ejes sobrantes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
# ---- 2️⃣ Multicolinealidad: Matriz de Correlación y VIF ---- #
# Matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(pd.DataFrame(x_train, columns=column_names).corr(), annot=True, cmap='inferno', linewidths=0.5)
plt.title("Matriz de Correlación", color="black")
plt.show()

# Cálculo del VIF (Factor de Inflación de Varianza)
vif_data = pd.DataFrame()
vif_data["Feature"] = column_names
vif_data["VIF"] = [variance_inflation_factor(x_train, i) for i in range(len(column_names))]

print("\n📌 **Factor de Inflación de Varianza (VIF):**")
print(vif_data)

# ---- 3️⃣ Independencia de Errores: Prueba Durbin-Watson ---- #
dw_stat = sm.stats.stattools.durbin_watson(logit)
print(f"\n📌 **Durbin-Watson Statistic:** {dw_stat:.3f}")
if dw_stat < 1.5 or dw_stat > 2.5:
    print("⚠️ Posible problema de autocorrelación.")
else:
    print("✅ No hay indicios de autocorrelación.")

# ---- 4️⃣ Outliers e Influencia: Distancia de Cook ---- #
# Ajustar modelo con StatsModels para calcular residuos
x_train_sm = sm.add_constant(x_train)  
logit_model = sm.Logit(y_train, x_train_sm)
result = logit_model.fit()

# Calcular influencia y distancia de Cook
influence = result.get_influence()
cooks_d = influence.cooks_distance[0]

# Graficar la Distancia de Cook
plt.figure(figsize=(10, 5))
plt.stem(cooks_d, markerfmt="ro", basefmt="b-")
plt.axhline(y=4 / len(x_train), color="black", linestyle="--", label="Umbral (4/N)")
plt.xlabel("Índices de Observación")
plt.ylabel("Distancia de Cook")
plt.title("Detección de Outliers con Distancia de Cook", color="black")
plt.legend()
plt.show()

# Identificar posibles outliers
outliers = np.where(cooks_d > (4 / len(x_train)))[0]
print(f"\n📌 **Observaciones con alta influencia (Outliers):** {outliers if len(outliers) > 0 else 'No hay valores problemáticos.'}")
```

**Prueba t de Student**: Diferencia de edad entre pacientes con y sin cáncer de pulmón.  

Esta prueba nos ayuda a determinar si la edad promedio difiere significativamente entre pacientes con y sin cáncer, lo que podría indicar que la edad es un factor de riesgo importante.  

**Hipótesis Nula**: No hay diferencia significativa en la edad promedio entre pacientes con y sin cáncer de pulmón.  
**Hipótesis Alternativa**: Existe una diferencia significativa en la edad promedio entre ambos grupos.

```python
# 1. Prueba t de Student: Diferencia de edad entre pacientes con y sin cáncer de pulmón
age_cancer = data_[data_['LUNG_CANCER'] == 1]['AGE']
age_no_cancer = data_[data_['LUNG_CANCER'] == 0]['AGE']
t_stat, p_value_ttest = stats.ttest_ind(age_cancer, age_no_cancer, equal_var=False)
print(f'Prueba t: t-stat={t_stat}, p-valor={p_value_ttest}')
```

Como podemos ver, el p-valor dio como resultado un valor mayor que 0.05, por tanto, podemos decir que no hay suficiente evidencia en este conjunto de datos para afirmar que la edad difiere significativamente entre las perssonas que tienen cancer de pulmón y las que no.

# Prueba de Chi-cuadrado: Asociación entre síntomas y cáncer de pulmón.
Esta prueba permite evaluar si síntomas como tos, dificultad para respirar y fatiga están relacionados con la presencia de cáncer de pulmón, lo que ayuda en el diagnóstico temprano.


```python
# 2. Prueba de Chi-cuadrado: Asociación entre síntomas y cáncer de pulmón
symptoms = ['COUGHING', 'SHORTNESS OF BREATH', 'FATIGUE ']
p_values_symptoms = {}
for symptom in symptoms:
    contingency = pd.crosstab(data_[symptom], data_['LUNG_CANCER'])
    chi2, p_value, _, _ = stats.chi2_contingency(contingency)
    p_values_symptoms[symptom] = p_value
    print(f'Chi-cuadrado para {symptom}: chi2={chi2}, p-valor={p_value}')

```

Los resultados nos llevan a la conclusión de que ni la tos ni la falta de aire  son fuertes indicadores de cancer de pulmón, mientas que la fatiga sí lo es.

# Prueba Z para diferencia de proporciones: Fumar y cáncer de pulmón.
Esta prueba de hipótesis analiza si la proporción de fumadores es significativamente diferente entre quienes tienen cáncer y quienes no, ayudando a confirmar si el tabaquismo es un factor de riesgo.

**Hipótesis Nula** : No hay diferencia en la proporción de fumadores entre pacientes con y sin cáncer de pulmón.  
**Hipótesis Alternativa**: Existe una diferencia significativa en la proporción de fumadores entre los grupos.

```python
# 3. Prueba Z para diferencia de proporciones: Fumar y cáncer de pulmón
count = np.array([data_[data_['LUNG_CANCER'] == 1]['SMOKING'].sum(), data_[data_['LUNG_CANCER'] == 0]['SMOKING'].sum()])
nobs = np.array([data_[data_['LUNG_CANCER'] == 1]['SMOKING'].count(), data_[data_['LUNG_CANCER'] == 0]['SMOKING'].count()])
z_stat, p_value_ztest = proportions_ztest(count, nobs)
print(f'Prueba Z (fumar vs. cáncer de pulmón): z-stat={z_stat}, p-valor={p_value_ztest}')
```

Los resultados arrojan que no hay suficiente evidencia para afirmar que fumar está significativamente asociado al cancer de pulmón.

# Regresión logística: Factores de riesgo del cáncer de pulmón.

**Hipótesis Alternativa**: Al menos una de las variables independientes influye significativamente en la probabilidad de tener cáncer de pulmón.


```python
# 5. Regresión logística para evaluar factores de riesgo en cáncer de pulmón
independent_vars = ['AGE', 'SMOKING', 'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'FATIGUE ']
X = data_[independent_vars]
X = sm.add_constant(X)
Y = data_['LUNG_CANCER']
logit_model = sm.Logit(Y, X)
result = logit_model.fit()
print(result.summary())
```
